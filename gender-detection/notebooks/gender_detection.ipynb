```python
train = pd.read_csv('../data/train_data.csv')
test = pd.read_csv('../data/test_data.csv')
```

```python
train[(train['gender'] == 'man')]
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train['gender'] = le.fit_transform(train['gender'])
```

```python
train = train.dropna()
```

```python
test = test.dropna()
```

```python
# Create the followers_to_following_ratio feature
train['followers_to_following_ratio'] = train['follower_count'] / train['following_count']

# Handle division by zero
train['followers_to_following_ratio'] = train['followers_to_following_ratio'].replace([np.inf, -np.inf], 0).fillna(0)
```

```python
# Create the followers_to_following_ratio feature
test['followers_to_following_ratio'] = test['follower_count'] / test['following_count']

# Handle division by zero
test['followers_to_following_ratio'] = test['followers_to_following_ratio'].replace([np.inf, -np.inf], 0).fillna(0)
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Define bins
bins_follower = [-1, 0, 100, 500, 1000, 2500]
labels_follower = ['none', 'low', 'medium', 'high', 'very_high']

bins_following = [-1, 0, 100, 500, 1000, 2100]
labels_following = ['none', 'low', 'medium', 'high', 'very_high']

# Categorize the columns
train['follower_count_bin'] = pd.cut(train['follower_count'], bins=bins_follower, labels=labels_follower)
train['following_count_bin'] = pd.cut(train['following_count'], bins=bins_following, labels=labels_following)

 # Initialize the OrdinalEncoder
encoder = LabelEncoder()

train['following_count_bin'] = encoder.fit_transform(train['following_count_bin'])

train['follower_count_bin'] = encoder.fit_transform(train['follower_count_bin'])

```

```python
from sklearn.preprocessing import OrdinalEncoder

# Define bins
bins_follower = [-1, 0, 100, 500, 1000, 2500]
labels_follower = ['none', 'low', 'medium', 'high', 'very_high']

bins_following = [-1, 0, 100, 500, 1000, 2100]
labels_following = ['none', 'low', 'medium', 'high', 'very_high']

# Categorize the columns
test['follower_count_bin'] = pd.cut(test['follower_count'], bins=bins_follower, labels=labels_follower)
test['following_count_bin'] = pd.cut(test['following_count'], bins=bins_following, labels=labels_following)

 # Initialize the OrdinalEncoder
encoder = LabelEncoder()

test['following_count_bin'] = encoder.fit_transform(test['following_count_bin'])

test['follower_count_bin'] = encoder.fit_transform(test['follower_count_bin'])

```

```python
train.drop(['follower_count', 'following_count'], axis=1, inplace=True)
```

```python
test.drop(['follower_count', 'following_count'], axis=1, inplace=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf.fit(train.fullname + ' ' + train.username + ' ' + train.biography)

from xgboost import XGBClassifier
x_train = train.drop('gender', axis=1)
y_train = train['gender']
txt_model = XGBClassifier()
txt_model.fit(tfidf.transform(x_train.fullname + ' ' + x_train.username + ' ' + x_train.biography), y_train)
```

```python
gender_by_txt_features_test = txt_model.predict(tfidf.transform(test.fullname + ' ' + test.username + ' ' + test.biography))
test['gender_by_txt'] = gender_by_txt_features_test
```

```python
gender_by_txt_features =  txt_model.predict(tfidf.transform(train.fullname + ' ' + train.username + ' ' + train.biography))
```

```python
data = train[['age', 'follower_count_bin', 'following_count_bin' , 'is_business', 'is_verified', 'is_private', 'gender']]
data['gender_by_txt'] = gender_by_txt_features
```

```python
test = test[['age', 'follower_count_bin', 'following_count_bin' , 'is_business', 'is_verified', 'is_private', 'gender_by_txt']]
```

```python
X = data.drop('gender', axis=1)
y = data['gender']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score


model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
f1_score(y_test, y_pred)
```

```python
import xgboost as xgb
from sklearn.model_selection import  GridSearchCV
from sklearn.metrics import f1_score


xgb = xgb.XGBClassifier()
xgb.fit(X_train, y_train)
y_pred = xgb.predict(X_test)
f1_score(y_test, y_pred)
```
